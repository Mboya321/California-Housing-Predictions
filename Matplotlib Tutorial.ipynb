```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
%matplotlib inline
```

```python
#creating a simple figure
fig, ax= plt.subplots()
plt.plot([1,2, 3, 4], [1,10, 5, 2]) #remember the x, y values mustv have the same dimensions
                                    #otherwise it will give you an error
```

```python
#adding more features
x= np.linspace(0, 2, 100) 
fig, ax= plt.subplots()  #creating an empty fig
ax.plot(x, x, label= 'linear')
ax.plot(x, x**2, label= 'quadratic')
ax.plot(x, x**3, label= 'cubic')
ax.set_xlabel('x_label')  #setting label for the x_axis
ax.set_ylabel('y_label')
ax.set_title('simple graph') #setting the title of the graph
ax.legend()  #setting the legend
```

```python
x= [1, 2, 3,5,6]
y= [0, 5, 6, 8, 7]
plt.plot(x, y)  #plots the graph with the default line and colour
```

```python
plt.plot(x, y, 'bo')  #plots the graph using the blue circle markers
```

```python
#make it more even beatiful
plt.plot(x, y, color= 'blue', linestyle= 'dashed', marker= 'o', linewidth= 2)
```

```python
#lets change the markers
plt.plot(x, y, color= 'magenta', linestyle= 'dashed', marker= '*', linewidth= 2)
```

```python
#plotting categorical data
groups= ['group_a', 'group_b', 'group_c']
values= [1, 5, 100]

plt.figure(figsize= (10, 5))  #setting the size of the figure

plt.subplot(131)   #create a graph for the bargraph
plt.bar(groups, values)  #plots a bar chart
plt.subplot(132)  #creates a graph for the scattering chart
plt.scatter(groups, values)  #scatters the data
plt.plot(groups, values)
plt.suptitle('plotting categorical data')
```

```python
#plotting histograms using the hist() function
housing= pd.read_csv(r'E:\AI\housing.csv')
housing.hist(bins= 50, figsize= (20,10))
```

```python
#making bar plots
data= [10, 45, 5, 56]
mp.pyplot.bar(data, height= 3)
```

```python
#making pie charts
data= [120000, 45000, 780000, 450000, 670000]
labels= ['food','entertainment', 'luxury', 'electricity', 'wife']
mp.pyplot.pie(data, labels= labels, shadow= 'True')
```

```python
#using the scatter() method
#the cmap param sets the color of the graph
#c param sets the range of colors for the graph
#alpha param sets the difference btn opaque and transparent areas
housing.plot.scatter(x= 'longitude', y= 'latitude', cmap= plt.get_cmap('jet'),
                    alpha= 0.6, label= 'population', c= 'median_house_value', colorbar= True)
```
